```python
import pandas as pd # data manipulation, preprocessing
```

```python
file_path = "/content/homicideinenglandandwalesappendixtablesyemar2024.xlsx"  # creating a var instead of directly loading it so it can be loaded with a specific table
```

```python
df = pd.read_excel(file_path, sheet_name='Table 1', skiprows=7) # sheet selection and skipping the top few rows ie filtering them out
```

```python
df.head(10) # sampling the data to make sure its been ingested and filtered accordingly
```

```python
df.replace('[x]', pd.NA, inplace=True) # the [x] boxes are better being converted to just NA so theres no error
```

```python
df.head() # sampling to see if the change has been made
```

```python
# TABLES NEEDED FOR ANALYSIS

# TABLE 20 UNDER INFLUENCE                                 DONE
# TABLE 22 DRUG RELATED DEALER ETC                         DONE
# TABLE 9 KNIFE CRIME AND TYPE OF KNIFE                    DONE
# TABLE 10 CRIME WITH TEENAGERS                            DONE
# TABLE 11 KNIFE CRIME 2014 TO 2024                        DONE
# TABLE 13 FIREARM CRIME                                   DONE
# TABLE 15 CRIME 16 AND OVER                               DONE
# TABLE 16 CRIME 16 AND UNDER                              DONE
```

```python
df_knifecrime_typeknife = pd.read_excel(file_path, sheet_name='Table 9', skiprows=6) # repeating same code but for another table with rows skipped based on the redundant rows in the flatfile
```

```python
df_teen_crime = pd.read_excel(file_path, sheet_name='Table 10', skiprows=6) # skiprows make sure the redundant data isnt there
```

```python
df_knife_crime2014_to24 = pd.read_excel(file_path, sheet_name='Table 11', skiprows=6)
```

```python
df_firearm_crime = pd.read_excel(file_path, sheet_name='Table 13', skiprows=6)
```

```python
table15a = pd.read_excel(file_path, sheet_name='Worksheet 15', skiprows=8, nrows=40) # using n rows to capture the right cells needed since a and b are stacked
table15b = pd.read_excel(file_path, sheet_name='Worksheet 15', skiprows=50, nrows=40)

# merging here is needed because 15a is the count and 15b is the percetnage so the tables should be joined using inner to retain all information as default
crime_age16_over = table15a.merge(
    table15b,
    on=['Sex of victim', 'Victim acquainted with suspect', 'Relationship to principal suspect'], how='inner',suffixes=('_count', '_percent'))
```

```python
table16a = pd.read_excel(file_path, sheet_name='Worksheet 16', skiprows=8, nrows=34) # using n rows to capture the right cells needed since a and b are stacked
table16b = pd.read_excel(file_path, sheet_name='Worksheet 16', skiprows=35, nrows=24)

# merging here is needed as before
crime_age16_under = table16a.merge(
    table16b,
    on=['Sex of victim', 'Victim acquainted with suspect', 'Relationship to principal suspect'], how='inner',suffixes=('_count', '_percent'))
```

```python
crime_underinfluence = pd.read_excel(file_path, sheet_name='Table 20', skiprows=6)
```

```python
drug_crime_related = pd.read_excel(file_path, sheet_name='Table 22', skiprows=6)
```

```python
crime_age16_under.head() # displaying a sample for the selected table to make sure theyre all correctly processed
```

```python
# saving the datasets now as seperate csv files to be implemented to streamlit
crime_age16_over.to_csv(r'/content/Crime_Aged 16+.csv', index=False)
df_knifecrime_typeknife.to_csv(r'/content/Knife crime with type of knife reported.csv', index=False)
df_teen_crime.to_csv(r'/content/Crime involving Teenagers.csv', index=False)
df_knife_crime2014_to24.to_csv(r'/content/Knife crime 2014 - 2024.csv', index=False)
df_firearm_crime.to_csv(r'/content/Crime involving Firearm.csv', index=False)
crime_age16_under.to_csv(r'/content/Crime_Aged under 16.csv', index=False)
crime_underinfluence.to_csv(r'/content/Crime under influence.csv', index=False)
drug_crime_related.to_csv(r'/content/Drug Related Crime.csv', index=False)
```
